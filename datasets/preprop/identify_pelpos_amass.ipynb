```python
paths = []
folders = []
dataset_names = []
for root, dirs, files in os.walk('./amass_data'):
    print(root, dirs, files)
#     for folder in dirs:
#         folders.append(os.path.join(root, folder))
    folders.append(root)
    for name in files:
        dataset_name = root.split('/')[1]
        if dataset_name not in dataset_names:
            dataset_names.append(dataset_name)
        paths.append(os.path.join(root, name))
```

```python
save_root = './id_pel_pos'
save_folders = [folder.replace('./amass_data', './id_pel_pos') for folder in folders]
for folder in save_folders:
    os.makedirs(folder, exist_ok=True)
group_path = [[path for path in paths if name in path] for name in dataset_names]
```

```python
group_path = group_path
all_count = sum([len(paths) for paths in group_path])
cur_count = 0
print(len(group_path))
```

```python
## --------------------------------------------------------------------------------------------------------------------
## -------------데이터셋의 Global Position과 pelvis의 데이터를 0으로 초기화----------------
## --------------------------------------------------------------------------------------------------------------------

import tqdm


# tqdm source: https://zephyrus1111.tistory.com/305
for paths in tqdm.tqdm(group_path, 
              total = len(group_path), ## 전체 진행수
              desc = 'Complete', ## 진행률 앞쪽 출력 문장
              ncols = 70, ## 진행률 출력 폭 조절
              ascii = ' =', ## 바 모양, 첫 번째 문자는 공백이어야 작동
              leave = True, ## True 반복문 완료시 진행률 출력 남김. False 남기지 않음.
             ):
    for path in paths:
        file_name = path
        # print(file_name)
        npz_data = np.load(file_name, allow_pickle=True)
        keys = list(npz_data.keys())
        # print(keys) 
            # ['trans', 'gender', 'mocap_framerate', 'betas', 'dmpls', 'poses']
            # ['gender', 'betas']
            # ['poses', 'gender', 'mocap_framerate', 'betas', 'marker_data', 'dmpls', 'marker_labels', 'trans']


        # rnd_int = random.randint(30,60)
        
        # dmpls = npz_data['dmpls']
        # poses = npz_data['poses']
        # print(dmpls.shape)
        # print(poses.shape)

        if  len(keys) == 2:
            gender = npz_data["gender"]
            betas = npz_data["betas"]
            save_path = file_name.replace("./amass_data","./id_pel_pos" )
            # print(save_path)
            np.savez(save_path,gender =gender, betas = betas )

        #######################################################################################################
        elif len(keys) == 6:
            trans = npz_data["trans"]
            gender = npz_data["gender"]
            mocap_framerate = npz_data["mocap_framerate"]
            betas = npz_data["betas"]
            dmpls = npz_data["dmpls"]
            poses = npz_data["poses"]
            print(gender)
            print(type(gender))

            print(f'betas.shape: {betas.shape}')
            print(f'betas: {betas}')
            print(f'dmpls.shape: {dmpls.shape}')

            # print(trans.shape)
            trans[:] = 0 # 포지션 값을 0으로 변경해서 자리에서 움직이지 않게 하기
            # print(trans[:])

            # print(poses.shape)
            poses[: ,0:3] = 0 #pelvis 각도를 0으로 두어서 모두가 서 있는 포즈에서 진행할 수 있게 하기
            # print(poses[: ,0:3])
            

            save_path = file_name.replace("./amass_data","./id_pel_pos" ) # 수정 필요
            # print(save_path)
            np.savez(save_path,trans =trans, gender = gender, mocap_framerate=mocap_framerate,
                      betas =betas, dmpls = dmpls, poses = poses)
        #######################################################################################################


        elif len(keys) == 8:
            poses = npz_data["poses"]
            gender = npz_data["gender"]
            mocap_framerate = npz_data["mocap_framerate"]
            betas = npz_data["betas"]
            marker_data = npz_data["marker_data"]
            dmpls = npz_data["dmpls"]
            marker_labels = npz_data["marker_labels"]
            trans = npz_data["trans"]

            
            trans[:] = 0 # 포지션 값을 0으로 변경해서 자리에서 움직이지 않게 하기
            poses[: ,0:3] = 0 #pelvis 각도를 0으로 두어서 모두가 서 있는 포즈에서 진행할 수 있게 하기
            # print(trans[:])
            # print(poses[: ,0:3])


            save_path = file_name.replace("./amass_data","./id_pel_pos" )
            # print(save_path)
            np.savez(save_path,poses =poses, gender = gender, mocap_framerate=mocap_framerate,
                      betas =betas, marker_data = marker_data, dmpls = dmpls, marker_labels=marker_labels,trans=trans)
        break
    break

```
